```python
# class ChatBot:
#   #Potential Negative response
#   negative_responses = ("no","nope","nah","naw","not a chance","sorry")
#   #Exit conversation keywords
#   exit_commands = ("quit","pause","exit","goodbye","bye","later")
#   #random starter questions
#   random_questions = (
#       "Why are you here?",
#       "Are there many humans like you?",
#       "What do you consume for sustenance?",
#       "Is there intelligent life on this planet?",
#       "Does Earth have a leader?",
#   )

#   def __init__(self):
#     self.alienbabble = {'describe_planet_intent': r'.*\s*your planet.*',
#                         'answer_why_intent': r'why\sare.*',
#                         'about_eliza': r'.*\s*eliza',
#                         'about_session': r'.*\s*session'
#                    }

#   def greet(self):
#     self.name = input("What is your name?\n")
#     will_help = input(
#         f"Hi {self.name}, I'm Chat-Bot. Will you help me learn about your planet?\n")
#     if will_help in self.negative_responses:
#       print("Ok, have a nice Earth day!")
#       return
#     self.chat()

#     def make_exit(self, reply):
#       for command in self.exit_commands:
#         if reply == command:
#           print("Okay, have a nice Earth day!")
#           return True
#     def chat(self):
#         reply = input(random.choice(self.random_questions)).lower
#         while not self.make_exit(reply):
#           reply = input(self.match_reply(reply))


#     def match_reply(self, reply):
#       for key, value in self.alienbabble.items():
#           intent = key
#           regex_pattern = value
#           found_match = re.match(regex_pattern, reply)
#           if found_match and intent == 'describe_planet_intent':
#                return self.describe_planet_intent()
#           elif found_match and intent == 'answer_why_intent':
#               return self.answer_why_intent()
#           elif found_match and intent == 'about_eliza':
#               return self.about_eliza()
#           elif found_match and intent == 'about_session':
#               return self.about_session()
#       if not found_match:
#           return self.no_match_intent()

#     def answer_why_intent(self):
#         responses = ("I come in peace\n", "I am here to collect data on your planet and its inhabitants\n",
#                       "I headr the coffee is good\n")
#         return random.choice(responses)

#         def about_eliza(self):
#             responses = ("I Eliza the world's first bot\n", "I was built by Late.Joseph Wenzbaum\n",
#                          "I am here to clarify your requests and doubts")
#             return random.choice(responses)

#         def about_session(self):
#           responses = ("Session is on as you open the bot\n",'Session was Cool!!')
#           return random.choice(responses)

#         def no_match_intent(self):
#             responses = (
#             "Please tell me more.\n", "Tell me more!\n", "Why do you say that?\n", "I see. Can you elaborate?\n",
#             "Interesting. Can you tell me more?\n", "I see. How do you think?\n", "Why?\n",
#             "How do you think I feel when you say that?\n")
#             return random.choice(responses)


# AlienBot = ChatBot()
# AlienBot.greet()

#as we provide at that time reply will generate as:-
# What is your name?
# Sakshi
# Hi Sakshi, I'm Chat-Bot. Will you help me learn about your planet?
# no
# Ok, have a nice Earth day!
class ChatBot:  # Defines the ChatBot class.
    # Potential negative responses from the user.
    negative_responses = ("no", "nope", "nah", "naw", "not a chance", "sorry")
    # Keywords to exit the conversation.
    exit_commands = ("quit", "pause", "exit", "goodbye", "bye", "later")
    # Random starter questions for the chatbot.
    random_questions = (
        "Why are you here?",
        "Are there many humans like you?",
        "What do you consume for sustenance?",
        "Is there intelligent life on this planet?",
        "Does Earth have a leader?",
    )

    def __init__(self):  # Constructor to initialize the chatbot.
        # Dictionary to store patterns and their corresponding intents.
        self.alienbabble = {
            'describe_planet_intent': r'.*\s*your planet.*',
            'answer_why_intent': r'why\sare.*',
            'about_eliza': r'.*\s*eliza',
            'about_session': r'.*\s*session'
        }

    def greet(self):  # Method to greet the user and start the conversation.
        self.name = input("What is your name?\n")  # Gets the user's name.
        will_help = input(
            f"Hi {self.name}, I'm Chat-Bot. Will you help me learn about your planet?\n")  # Asks for user's consent.
        if will_help in self.negative_responses:  # Checks if the user declined.
            print("Ok, have a nice Earth day!")  # Prints a farewell message.
            return  # Exits the greet method if the user declined.
        self.chat()  # Starts the chat if the user agreed.

    def make_exit(self, reply):  # Method to check if the user wants to exit.
        for command in self.exit_commands:  # Iterates through exit commands.
            if reply == command:  # Checks if the user's reply matches an exit command.
                print("Okay, have a nice Earth day!")  # Prints a farewell message.
                return True  # Returns True to indicate exit.

    def chat(self):  # Method to handle the main chat loop.
        reply = input(random.choice(self.random_questions)).lower()  # Asks a random question and converts the reply to lowercase.
        while not self.make_exit(reply):  # Continues the loop until the user wants to exit.
            reply = input(self.match_reply(reply))  # Gets the next reply based on the previous one.

    def match_reply(self, reply):  # Method to match the user's reply to an intent.
        for key, value in self.alienbabble.items():  # Iterates through intents and patterns.
            intent = key  # Assigns the current key to intent.
            regex_pattern = value  # Assigns the current value to regex_pattern.
            found_match = re.match(regex_pattern, reply)  # Checks if the reply matches the pattern.
            if found_match and intent == 'describe_planet_intent':  # If matched, calls the corresponding method.
                return self.describe_planet_intent()
            elif found_match and intent == 'answer_why_intent':
                return self.answer_why_intent()
            elif found_match and intent == 'about_eliza':
                return self.about_eliza()
            elif found_match and intent == 'about_session':
                return self.about_session()
        if not found_match: # if no intent matched.
            return self.no_match_intent() #returns a default response

    def answer_why_intent(self):  # Method to answer "why" questions.
        responses = ("I come in peace\n", "I am here to collect data on your planet and its inhabitants\n",
                      "I heard the coffee is good\n")
        return random.choice(responses)  # Returns a random response.

    def about_eliza(self):  # Method to provide information about Eliza.
        responses = ("I Eliza the world's first bot\n", "I was built by Late.Joseph Wenzbaum\n",
                         "I am here to clarify your requests and doubts")
        return random.choice(responses)  # Returns a random response.

    def about_session(self):  # Method to provide information about the session.
        responses = ("Session is on as you open the bot\n", 'Session was Cool!!')
        return random.choice(responses)  # Returns a random response.

    def no_match_intent(self):  # Method to handle unmatched replies.
        responses = (
            "Please tell me more.\n", "Tell me more!\n", "Why do you say that?\n", "I see. Can you elaborate?\n",
            "Interesting. Can you tell me more?\n", "I see. How do you think?\n", "Why?\n",
            "How do you think I feel when you say that?\n")
        return random.choice(responses)  # Returns a random response.

    def describe_planet_intent(self): # Method to handle planet description intent.
        responses = ("My planet is a beautiful place, filled with lush greenery and crystal-clear waters.\n", "It's a vibrant world with diverse ecosystems and fascinating creatures.\n")
        return random.choice(responses) # Returns a random response.


AlienBot = ChatBot()  # Creates an instance of the ChatBot class.
AlienBot.greet()

# Output:-
# What is your name?
# Sakshi
# Hi Sakshi, I'm Chat-Bot. Will you help me learn about your planet?
# Why are you here?
# What do you consume for sustenance?I heard the coffee is good
# I see. Can you elaborate?
# Please tell me more
# How do you think I feel when you say that?
# My planet is a beautiful place, filled with lush greenery and crystal-clear waters.
# I see. Can you elaborate?
# I was built by Late.Joseph Wenzbaum
# Interesting. Can you tell me more?
# good planet
# How do you think I feel when you say that?
# nice
# I see. Can you elaborate?
# yeah what should I say no words to secribe it
# I see. Can you elaborate?
# quit
# Okay, have a nice Earth day!


```
